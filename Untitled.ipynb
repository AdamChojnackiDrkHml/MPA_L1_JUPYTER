```python
import random
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
import numpy as np
import math
```

# Produkt

```python
class Product:
    def __createAnomaly(self, value):
        return value * (10 **random.randint(-2, 2))
    
    def __createPrices(self, sampleSize, minPrice, maxPrice, chanceOfAnomaly):
        currPrice = minPrice
        totalPriceChange = maxPrice - minPrice
        priceDelta = totalPriceChange / sampleSize
        
        prices = list(range(sampleSize))
        
        for i in range(len(prices)):
            prices[i] = currPrice + (random.uniform(-0.2, 0.2) * totalPriceChange)
            
            if random.random() < chanceOfAnomaly:
                prices[i] = self.__createAnomaly(prices[i])
            
            currPrice = currPrice + priceDelta
    
        return prices
    
    def __createAmounts(self, sampleSize, minAmount, maxAmount, chanceOfAnomaly):
        amounts = list(range(sampleSize))
        
        for i in range(len(amounts)):
            amounts[i] = random.randint(minAmount, maxAmount)
                
            if random.random() < chanceOfAnomaly:
                amounts[i] = self.__createAnomaly(amounts[i])
        
        return amounts
    
    def __init__(self, sampleSize, minPrice, maxPrice, minAmount, maxAmount, chanceOfAnomaly, name):
        self.prices = self.__createPrices(sampleSize, minPrice, maxPrice, chanceOfAnomaly)
        self.amounts = self.__createAmounts(sampleSize, minAmount, maxAmount, chanceOfAnomaly)
        self.name = name
    

            
        
```

```python
exampleProduct = Product(50, 100, 124, 20, 40, 0.2, "Parasol")

productIndexes = [i for i in range(50)]
productPrices = [price for price in exampleProduct.prices]
productAmounts = [amount for amount in exampleProduct.amounts]
```

# Baza danych

```python
class Database:
    def __init__(self, sampleSizes, productNames, pricesMatrix, amountsMatrix, chanceOfAnomaly):
        self.products = dict()
        
        for i in range(len(sampleSizes)):
            self.products[productNames[i]] = Product(
                sampleSizes[i],
                pricesMatrix[i][0],
                pricesMatrix[i][1], 
                amountsMatrix[i][0], 
                amountsMatrix[i][1], 
                chanceOfAnomaly, 
                productNames[i]
            )
```

```python
database = Database([50], ["Latarka"], [[100, 124]], [[40, 60]], 0.1)
```

# Obróbka danych

```python
def linearRegression(indexes, values):
    indexes_np = np.array(indexes). reshape((-1,1))
    values_np = np.array(values)
    model = LinearRegression().fit(indexes_np, values_np)
    
    return (model.coef_[0], model.intercept_)
```

```python
def printDataWithLinearRegression(indexes, values, coefficients):
    plt.scatter(indexes, values)
    linearRegressionXValues = [0, len(indexes)]
    linearRegressionYValues = [coefficients[1], coefficients[0] * len(indexes) + coefficients[0]]
    plt.plot(linearRegressionXValues, linearRegressionYValues)
    plt.show()
    plt.clf()
```

```python
coeafficientsLRPrice = linearRegression(productIndexes, productPrices)
printDataWithLinearRegression(productIndexes, productPrices, coeafficientsLRPrice)
```

```python
coeafficientsLRAmounts = linearRegression(productIndexes, productAmounts)
printDataWithLinearRegression(productIndexes, productAmounts, coeafficientsLRAmounts)
```

```python
def removeOutliers(indexes, values, coefficients):
    expectedValues = [coefficients[0] * i + coefficients[1] for i in range(len(indexes))]

    SSE = sum([(values[i] - expectedValues[i]) ** 2 for i in range(len(indexes))])
    
    s = math.sqrt(SSE / (len(indexes)))
                  
    goodValues = list()
    removed = 0
                  
    for i in range(len(indexes)):
        if abs(values[i] - expectedValues[i]) < abs(2 * s):
            goodValues.append((i, values[i]))
        else:
            removed += 1
    
    unzipped = list(zip(*goodValues))
    
    return (unzipped[0], unzipped[1], removed)
    
```

```python
indexesC, pricesC, removed = removeOutliers(productIndexes, productPrices, coeafficientsLRPrice)
plt.scatter(indexesC, pricesC)
plt.show()
plt.clf()
```

```python
indexesAmountsC, amountsC, removed = removeOutliers(productIndexes, productAmounts, coeafficientsLRPrice)
plt.scatter(indexesAmountsC, amountsC)
plt.show()
plt.clf()
```

### Chcemy powtórzyć usuwanie punktów odstających (tzw dużej dźwigni)

```python
def removeOutliersUntilPure(indexes, values):
    while True:
        coeffsLR = linearRegression(indexes, values)
        indexes, values, removed = removeOutliers(indexes, values, coeffsLR)
        
        if removed == 0:
            return (indexes, values)
```

```python
indexesC, pricesC = removeOutliersUntilPure(productIndexes, productPrices)
plt.scatter(indexesC, pricesC)
plt.show()
plt.clf()
```

```python
indexesAmountsC, amountsC = removeOutliersUntilPure(productIndexes, productAmounts)
plt.scatter(indexesAmountsC, amountsC)
plt.show()

plt.clf()
```

```python
def printWholeDatabase(database):
    for name, product in database.products.items():
        plt.scatter(range(len(product.prices)), product.prices, label = name)
    
    plt.legend(loc="upper right")
    plt.title("Prices")
    plt.show()
    plt.clf()
    
    for name, product in database.products.items():
        plt.scatter(range(len(product.prices)), product.amounts, label = name)
    
    plt.legend(loc="upper right")
    plt.title("Amounts")
    plt.show()
    plt.clf()
    
    for name, product in database.products.items():
        i, p = removeOutliersUntilPure(range(len(product.prices)), product.prices)
        plt.scatter(i, p, label = name)
    
    plt.legend(loc="upper right")
    plt.title("Prices cleared")
    plt.show()
    plt.clf()
    
    for name, product in database.products.items():
        i, a = removeOutliersUntilPure(range(len(product.prices)), product.amounts)
        plt.scatter(i, a, label = name)
    
    plt.legend(loc="upper right")
    plt.title("Amounts cleared")
    plt.show()
    plt.clf()    
```

```python
productNames = [
    "Latarka",
    "Parasol", 
    "Tulipan",
]

rangesOfPrices = [
    [100, 124],
    [200, 140],
    [19, 40],
]

rangesOfAmounts = [
    [40, 60],
    [10, 30],
    [100, 300],
]

sampleSizes = [
    50,
    50, 
    50,
]

mutliProductDB = Database(sampleSizes, productNames, rangesOfPrices, rangesOfAmounts, 0.1)

printWholeDatabase(mutliProductDB)
```

# Predykator

```python
class Predictor:
    
    def differencesFromLR(self, indexes, values, coeffs):
        expectedValues = [coeffs[0] * i + coeffs[1] for i in range(len(indexes))]
        return [values[i] - expectedValues[i] for i in range(len(indexes))]
    
    def markov(self, diffs, sensitibility):
        EXdiff = sum([abs(diff) for diff in diffs]) / (len(diffs) - 1)
        return EXdiff / sensitibility
    
    def chebyshev(self, diffs, sensitibility):
        EXdiff = sum([abs(diff) for diff in diffs]) / (len(diffs)- 1)
        VarXdiff = sum([(diff - EXdiff) ** 2 for diff in diffs]) / (len(diffs) - 1)
        return math.sqrt(VarXdiff/ sensitibility)
    
    def printThresholdForData(self, indexes, values, thresholdDiff, coeffs):
        plt.scatter(indexes, values)
        plt.plot([0, len(indexes)], [coeffs[1] + thresholdDiff, len(indexes) * coeffs[0] + coeffs[1] + thresholdDiff], label = "upper bound")
        plt.plot([0, len(indexes)], [coeffs[1], len(indexes) * coeffs[0] + coeffs[1]])
        plt.plot([0, len(indexes)], [coeffs[1] - thresholdDiff, len(indexes) * coeffs[0] + coeffs[1] - thresholdDiff], label = "lower bound")
        plt.legend()
        plt.show()
        plt.clf()
    
    def quickSketchForProduct(self, product):
        prices = product.prices
        amounts = product.amounts
        indexes = [i for i in range(len(prices))]
        
        
        indexesC, pricesC = removeOutliersUntilPure(indexes, prices)
        indexesAC, amountsC = removeOutliersUntilPure(indexes, amounts)
        
        coeffsPrice = linearRegression(indexesC, pricesC)
        coeffsAmount = linearRegression(indexesAC, amountsC)
    
        diffsPrice = self.differencesFromLR(indexesC, pricesC, coeffsPrice)
        diffsAmounts = self.differencesFromLR(indexesAC, amountsC, coeffsAmount)
        
        markovTP = self.markov(diffsPrice, 0.6)
        chebyshevTP = self.chebyshev(diffsPrice, 0.6)
        
        markovTA = self.markov(diffsPrice, 0.6)
        chebyshevTA = self.chebyshev(diffsPrice, 0.6)
        
        self.printThresholdForData(indexesC, pricesC, markovTP, coeffsPrice)
        self.printThresholdForData(indexesC, pricesC, chebyshevTP, coeffsPrice)
        
        self.printThresholdForData(indexesAC, amountsC, markovTA, coeffsAmount)
        self.printThresholdForData(indexesAC, amountsC, chebyshevTA, coeffsAmount)
        
    def checkThreshold(self, value, expectedValue, threshold):
        return abs(value - expectedValue) < threshold
    
    def checkAll(self, newValue, coeffs, newIndex, diffs):
        markovT = self.markov(diffs, 0.6)
        chebyshevT = self.chebyshev(diffs, 0.6)
                
        expectedValue = newIndex * coeffs[0] + coeffs[1]
        
        markovTest = self.checkThreshold(newValue, expectedValue, markovT)
        chebyshevTest = self.checkThreshold(newValue, expectedValue, chebyshevT)
#         print(newValue, expectedValue, f"Markov - {markovT}", f"Test - {markovTest}")
#         print(newValue, expectedValue, f"Chebyshev - {chebyshevT}", f"Test - {chebyshevTest}")
        if not markovTest or not chebyshevTest:
            print(f"Adding of value {newValue} failed")
            print(f"Markov Test: {markovTest}, should be in range [{expectedValue - markovT}, {expectedValue + markovT}]")
            print(f"Chebyshev Test: {chebyshevTest}, should be in range [{expectedValue - chebyshevT}, {expectedValue + chebyshevT}]")
        
        return markovTest and chebyshevTest 
    
    def addData(self, indexes, prices, amounts, newDataPrice, newDataAmount):
        indexesC, pricesC = removeOutliersUntilPure(indexes, prices)
        indexesAC, amountsC = removeOutliersUntilPure(indexes, amounts)
        
        coeffsPrice = linearRegression(indexesC, pricesC)
        coeffsAmount = linearRegression(indexesAC, amountsC)
        
        diffsPrice = self.differencesFromLR(indexesC, pricesC, coeffsPrice)
        diffsAmounts = self.differencesFromLR(indexesAC, amountsC, coeffsAmount)
        newIndex = len(indexesC)
        
        resultPrice = self.checkAll(newDataPrice, coeffsPrice, newIndex, diffsPrice)
        resultAmount = self.checkAll(newDataAmount, coeffsAmount, newIndex, diffsAmounts)
        
        if resultPrice and resultAmount:
            indexes.append(newIndex)
            prices.append(newDataPrice)
            amounts.append(newDataAmount)
            return (prices, amounts, True)
        else: 
            return (prices, amounts, False)
        
    def addItemToDatabase(self, database, productName, newProductPrice, newProductAmount):
        if productName not in database.products:
            print(f"{productName} is not in database")
            return False
    
        product = database.products[productName]
        
        indexes = [i for i in range(len(product.prices))]        
        prices, amounts, result = self.addData(indexes , product.prices, product.amounts, newProductPrice, newProductAmount)
        
        if result:
            product.prices = prices
            product.amounts = amounts
            database.products[productName] = product
            print(f"Product succesfully added: {productName} - ({newProductPrice}, {newProductAmount})")
            return True
        else:
            print("Failed to add product to database")
            return False
```

# Testowanie

```python
class Tester:
#     Currently only one product testin is allowed
    def test(self, items, productName, database):
        predictor = Predictor()
        predictor.quickSketchForProduct(database.products[productName])
        return [predictor.addItemToDatabase(database, productName, price, amount) for (price, amount) in items]
            
    def testBadItemPrices(self):
        rangeOfPrices = [100, 124]
        rangeOfAmounts = [40, 60]
        database = Database([50], ["Latarka"], [rangeOfPrices], [rangeOfAmounts], 0.1)
        
        badItemPrices = [random.randint(150, 200) for i in range(10)]
        badItemPrices2 = [random.randint(20, 80) for i in range(10)]
        badItemPrices = badItemPrices + badItemPrices2
        itemAmounts = [random.randint(50, 60) for i in range(20)]
        
        badItems = list(zip(badItemPrices, itemAmounts))
        
        result = self.test(badItems, "Latarka", database)
        
        if any(result):
            print("Well that should not happen")
            
    def testBadItemAmounts(self):
        rangeOfPrices = [100, 124]
        rangeOfAmounts = [40, 60]
        database = Database([50], ["Latarka"], [rangeOfPrices], [rangeOfAmounts], 0.1)
        
        baditemAmounts = [random.randint(70, 100) for i in range(10)]
        baditemAmounts2 = [random.randint(10, 30) for i in range(10)]
        baditemAmounts = baditemAmounts + baditemAmounts2
        itemPrices = [random.randint(50, 60) for i in range(20)]
        
        badItems = list(zip(itemPrices, baditemAmounts))
        
        result = self.test(badItems, "Latarka", database)    
        
        if any(result):
            print("Well that should not happen")
            
    def testDifferentTypesOfItems(self, productNames, items2D, database):

        for i, productName in enumerate(productNames):
            self.test(items2D[i], productName, database)
```

```python
tester = Tester()

tester.testBadItemPrices()
tester.testBadItemAmounts()
```
